```python
import pandas as pd
import sys
import matplotlib.pyplot as plt
from sklearn import preprocessing
from sklearn.linear_model import LinearRegression
```

```python
df = pd.read_csv(r"D:\TeachOneself\PythonWeb\MachineLearing\buoi_16_ThTon\population.csv")
df.head()
```

```python
df.isna().sum()
```

```python
df['Migrants(net)'] = df['Migrants(net)'].fillna(df['Migrants(net)'].mean())
df.isna().sum()
```

```python
preprocess = preprocessing.StandardScaler()
```

```python
X = df[["Year", "Median Age", "Fertility Rate",
        "Density (P/Km²)", "Urban Pop", "Migrants(net)"]]
y = df["Population"]
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=1)
```

```python
import seaborn as sns
sns.displot(df)
```

```python
lr = LinearRegression()
lr.fit(x_train, y_train)
```

```python
y_pred = lr.predict(x_test)
```

```python
from sklearn.metrics import r2_score, mean_absolute_error, mean_squared_error
mse = mean_squared_error(y_test, y_pred)
mae = mean_absolute_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)
```

```python
print(f"MAE: {mae}")
print(f"MSE: {mse}")
print(f"R2: {r2}")
```

```python
from sklearn.tree import DecisionTreeClassifier
cls_dt = DecisionTreeClassifier()
cls_dt.fit(x_train, y_train)
y_pred = cls_dt.predict(x_test)
from sklearn.metrics import r2_score, mean_absolute_error, mean_squared_error
mse = mean_squared_error(y_test, y_pred)
mae = mean_absolute_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)
print(f"MAE: {mae}")
print(f"MSE: {mse}")
print(f"R2: {r2}")
```
