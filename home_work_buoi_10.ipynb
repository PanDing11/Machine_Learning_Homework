```python
# Bài 1: Lớp con kế thừa lớp cha
# Mô tả: Tạo một lớp cha Animal với phương thức speak(). Tạo lớp con Dog kế thừa từ Animal và ghi đè (override) phương thức speak() để trả về âm thanh của chó.
# Yêu cầu:
# Lớp Animal có phương thức speak() trả về "I make a sound".
#
# Lớp Dog ghi đè phương thức speak() để trả về "Woof!".
# Đầu vào/Đầu ra:
# Gọi Dog().speak() → "Woof!"
#
# Gọi Animal().speak() → "I make a sound"
class Animal:
    def __init__(self, name:str, tell:str):
        self.name = name
        self.tell = tell


    def speak(self):
        print("I make a sound")

class Dog(Animal):
    def __init__(self, name:str, tell:str ):
        super().__init__(name, tell)
    def speak(self):
        print(f"{self.name} says {self.tell}, Woof!")

class Cat(Animal):
    def __init__(self, name:str, tell:str ):
        super().__init__(name, tell)
    def speak(self):
        print(f"{self.name} says {self.tell}, Woof!")
animal = Animal("Milo", "Hi")
animal.speak()

dog = Dog('Budy', 'gau')
dog.speak()

cat = Cat('Julia', 'meow')
cat.speak()
```

```python
# Bài 2: Kế thừa với thuộc tính
# Mô tả: Tạo lớp cha Vehicle với thuộc tính brand và phương thức info(). Tạo lớp con Car kế thừa từ Vehicle, thêm thuộc tính num_doors và ghi đè phương thức info() để hiển thị thông tin đầy đủ.
# Yêu cầu:
# Lớp Vehicle có thuộc tính brand và phương thức info() trả về "Brand: {brand}".

# Lớp Car có thêm thuộc tính num_doors và ghi đè info() trả về "Brand: {brand}, Number of doors: {num_doors}".
# Đầu vào/Đầu ra:
# Tạo Car("Toyota", 4).info() → "Brand: Toyota, Number of doors: 4"

class Vehicle:
    def __init__(self, brand:str):
        self.brand = brand

    def infor(self):
        return f"Brand: {self.brand}"

class Car(Vehicle):
    def __init__(self, brand:str, num_doors:int):
        super().__init__(brand)
        self.num_doors = num_doors

    def infor(self):
        return f"Brand: {self.brand}, Number of doors: {self.num_doors}"

car = Car("Toyota", 4)
car.infor()
```

```python
# 1. Lớp Vehicle
class Vehicle:
    def __init__(self, brand: str, model: str, year: int, weight: int, _is_runge: bool = False):
        self.brand = brand
        self.model = model
        self.year = year
        self.weight = weight
        self._is_runge = _is_runge

    # Phương thức cụ thể: khởi động động cơ
    def start_engine(self):
        self._is_runge = True
        return f"Động cơ khởi động cho {self.brand} {self.model}"

    # Phương thức cụ thể: tắt động cơ
    def stop_engine(self):
        self._is_runge = False
        return f"Động cơ đã tắt cho {self.brand} {self.model}"

    # Phương thức cụ thể: trả về thông tin cơ bản
    def get_info(self):
        return f"Thương hiệu: {self.brand}, Mẫu: {self.model}, Năm: {self.year}, Trọng lượng: {self.weight}, Đang chạy: {self._is_runge}"

    # Phương thức placeholder: tính hiệu suất nhiên liệu
    def calculate_fuel_efficiency(self, distance, fuel_used):
        return "Tính hiệu suất nhiên liệu chưa được triển khai cho Vehicle chung"

    # Phương thức placeholder: kiểm tra bảo dưỡng
    def maintenance_check(self):
        return "Kiểm tra bảo dưỡng chưa được triển khai cho Vehicle chung"

# 2. Lớp Car kế thừa từ Vehicle
class Car(Vehicle):
    def __init__(self, brand: str, model: str, year: int, weight: int, _is_runge: bool, num_doors: int, fuel_type: str):
        super().__init__(brand, model, year, weight, _is_runge)
        self.num_doors = num_doors
        self.fuel_type = fuel_type

    # Phương thức override: khởi động động cơ
    def start_engine(self):
        super().start_engine()  # Gọi phương thức cha để cập nhật _is_runge
        return f"Nhiên liệu đang dùng là {self.fuel_type}"

    # Phương thức override: tính hiệu suất nhiên liệu (km/l)
    def calculate_fuel_efficiency(self, distance, fuel_used):
        if distance < 0 or fuel_used < 0:
            return "Lỗi: Khoảng cách và nhiên liệu sử dụng phải không âm!"
        if fuel_used == 0:
            return "Lỗi: Nhiên liệu sử dụng không thể bằng 0!"
        return distance / fuel_used

    # Phương thức override: kiểm tra dầu và lốp
    def maintenance_check(self):
        return f"Kiểm tra dầu và lốp cho {self.brand} {self.model}"

# 3. Lớp Motorcycle kế thừa từ Vehicle
class Motorcycle(Vehicle):
    def __init__(self, brand: str, model: str, year: int, weight: int, _is_runge: bool, has_sidecar: bool):
        super().__init__(brand, model, year, weight, _is_runge)
        self.has_sidecar = has_sidecar

    # Nếu có sidecar, hiệu suất nhiên liệu giảm 20%
    def calculate_fuel_efficiency(self, distance, fuel_used):
        if distance < 0 or fuel_used < 0:
            return "Lỗi: Khoảng cách và nhiên liệu sử dụng phải không âm!"
        if fuel_used == 0:
            return "Lỗi: Nhiên liệu sử dụng không thể bằng 0!"
        efficiency = distance / fuel_used
        return efficiency * 0.8 if self.has_sidecar else efficiency

# 4. Lớp Truck kế thừa từ Vehicle
class Truck(Vehicle):
    def __init__(self, brand: str, model: str, year: int, weight: int, _is_runge: bool, cargo_capacity: float):
        super().__init__(brand, model, year, weight, _is_runge)
        self.cargo_capacity = cargo_capacity

    # Hiệu suất nhiên liệu giảm khi trọng tải lớn
    def calculate_fuel_efficiency(self, distance, fuel_used):
        if distance < 0 or fuel_used < 0:
            return "Lỗi: Khoảng cách và nhiên liệu sử dụng phải không âm!"
        if fuel_used == 0:
            return "Lỗi: Nhiên liệu sử dụng không thể bằng 0!"
        return (distance / fuel_used) / (1 + self.cargo_capacity / 1000)
# 5. Lớp ElectricCar kế thừa từ Car
class ElectricCar(Car):
    def __init__(self, brand: str, model: str, year: int, weight: int, _is_runge: bool,
                 num_doors: int, fuel_type: str, battery_capacity: float, _battery_level: float = 0):
        super().__init__(brand, model, year, weight, _is_runge, num_doors, fuel_type)
        self.battery_capacity = battery_capacity
        self._battery_level = min(_battery_level, battery_capacity)

    def charge_battery(self, charge_amount: float):
        if charge_amount < 0:
            return "Lỗi: Lượng sạc phải không âm!"
        self._battery_level = min(self._battery_level + charge_amount, self.battery_capacity)
        return f"{self.brand} {self.model} đã sạc đến {self._battery_level} kWh"

    def calculate_fuel_efficiency(self, distance, energy_used):
        if distance < 0 or energy_used < 0:
            return "Lỗi: Khoảng cách và năng lượng sử dụng phải không âm!"
        if energy_used == 0:
            return "Lỗi: Năng lượng sử dụng không thể bằng 0!"
        return distance / energy_used


# 6. Lớp HybridCar kế thừa từ Car và ElectricCar
class HybridCar(ElectricCar):  # sửa lại để kế thừa đúng logic
    def __init__(self, brand: str, model: str, year: int, weight: int, _is_runge: bool,
                 num_doors: int, fuel_type: str, battery_capacity: float,
                 _battery_level: float, current_mode: str = "Electric"):
        super().__init__(brand, model, year, weight, _is_runge, num_doors, fuel_type,
                         battery_capacity, _battery_level)
        self.current_mode = current_mode if current_mode in ["Electric", "Petrol"] else "Electric"

    def switch_mode(self, mode: str):
        if mode not in ["Electric", "Petrol"]:
            return "Lỗi: Chế độ không hợp lệ! Chọn 'Electric' hoặc 'Petrol'."
        self.current_mode = mode
        return f"{self.brand} {self.model} đã chuyển sang chế độ {self.current_mode}"

    def calculate_fuel_efficiency(self, distance, fuel_or_energy_used):
        if distance < 0 or fuel_or_energy_used < 0:
            return "Lỗi: Khoảng cách và nhiên liệu/năng lượng sử dụng phải không âm!"
        if fuel_or_energy_used == 0:
            return "Lỗi: Nhiên liệu/năng lượng sử dụng không thể bằng 0!"
        efficiency = distance / fuel_or_energy_used
        if self.current_mode == "Petrol":
            return efficiency * 0.9
        return efficiency

    def get_info(self):
        return f"{super().get_info()}, Chế độ: {self.current_mode}, Mức pin: {self._battery_level} kWh"

# 7. Lớp Bus kế thừa từ Vehicle
class Bus(Vehicle):
    def __init__(self, brand: str, model: str, year: int, weight: int, _is_runge: bool, max_passengers: int):
        super().__init__(brand, model, year, weight, _is_runge)
        self.max_passengers = max_passengers

    def start_engine(self):
        super().start_engine()
        return f"{self.brand} {self.model} bus engine started with max {self.max_passengers} passengers."

    def calculate_fuel_efficiency(self, distance, fuel_used):
        if distance < 0 or fuel_used < 0:
            return "Lỗi: Khoảng cách và nhiên liệu sử dụng phải không âm!"
        if fuel_used == 0:
            return "Lỗi: Nhiên liệu sử dụng không thể bằng 0!"
        return (distance / fuel_used) * (1 - min(self.max_passengers, 100) / 500)

    def maintenance_check(self):
        return f"Checking passenger seating and brakes for {self.brand} {self.model}."

    def get_info(self):
        return f"{super().get_info()}, Số hành khách tối đa: {self.max_passengers}"

# 8. Hàm kiểm tra
def test_vehicle_system(vehicles):
    for vehicle in vehicles:
        print(f"\n=== Kiểm tra {vehicle.__class__.__name__} ===")
        print("Thông tin:", vehicle.get_info())
        print("Khởi động động cơ:", vehicle.start_engine())

        if isinstance(vehicle, ElectricCar) or isinstance(vehicle, HybridCar):
            print("Hiệu suất (km/kWh):", vehicle.calculate_fuel_efficiency(100, 5))
        else:
            print("Hiệu suất (km/l):", vehicle.calculate_fuel_efficiency(100, 5))
        print("Kiểm tra bảo dưỡng:", vehicle.maintenance_check())
        # Kiểm tra các phương thức đặc biệt nếu có
        if isinstance(vehicle, ElectricCar) or isinstance(vehicle, HybridCar):
            print("Sạc pin (10 kWh):", vehicle.charge_battery(10))
        if isinstance(vehicle, HybridCar):
            print("Chuyển chế độ (Petrol):", vehicle.switch_mode("Petrol"))
            print("Hiệu suất ở chế độ Petrol (km/l):", vehicle.calculate_fuel_efficiency(100, 5))
            print("Chuyển chế độ (Electric):", vehicle.switch_mode("Electric"))
        print("Tắt động cơ:", vehicle.stop_engine())

vehicles = [
    Vehicle("Generic", "Model", 2010, 1000, False),
    Car("Toyota", "Camry", 2020, 1500, False, 4, "Gasoline"),
    Motorcycle("Honda", "CBR", 2018, 200, False, True),
    Truck("Volvo", "FH16", 2015, 5000, False, 2000),
    ElectricCar("Tesla", "Model 3", 2022, 1800, False, 4, "Electric", 75, 50),
    HybridCar("Toyota", "Prius", 2021, 1600, False, 4, "Hybrid", 50, 30),
    Bus("Mercedes", "Citaro", 2019, 12000, False, 80)
]
test_vehicle_system(vehicles)
```
